```python
import seaborn as sns
```

```python
sns.stripplot(x='Party', y='Number of Active Accounts', data=congress)
```

```python
sns.set_style('ticks')
sns.barplot(x='Party', y='Total Posts', data=congress)
sns.despine(left=True)
```

```python
sns.set_context('paper')
sns.set_style('whitegrid')
sns.barplot(x='Party', y='Max Total Followers', data=congress)
sns.despine(left=True)
```

```python
sns.lmplot(x="Number of Active Accounts", y="Max Total Followers", data=congress)
```

```python
sns.barplot(x="Max Total Followers", y="Start of Term", data=congress)
```

```python
sns.lmplot(x="Total Posts", y="Max Total Followers", data=congress)
```

```python
sns.catplot(x="Platform", y="Total Posts", hue="Platform",
            kind="violin", split=True, data=congress);
```
